```python
import requests
from bs4 import BeautifulSoup

```

```python
URL="https://www.flipkart.com/samsung-galaxy-s21-fe-5g-snapdragon-888-graphite-256-gb/p/itm766f7432b6626?pid=MOBGSXD7DXBFQEDE&lid=LSTMOBGSXD7DXBFQEDEV4VOVH&marketplace=FLIPKART&store=tyy%2F4io&srno=b_1_1&otracker=browse&fm=neo%2Fmerchandising&iid=0476138a-c642-4229-bbe1-8769946245fb.MOBGSXD7DXBFQEDE.SEARCH&ppt=clp&ppn=mobile-phones-store&ssid=5nmogaki740000001707979512129"
r = requests.get(URL) 
data = r.content
print(data)
```

STEP 2:- Parse the HTML

```python
Soup = BeautifulSoup(data,"html.parser")
Soup.prettify
```

STEP 3:- HTML tree traversal

    Commonly Used Types Of Objects:
    
    (1) Tag
    (2) Navigablestring
    (3) Beautifulsoup
    (4) Comment

(1) Check The Product Price:-

```python
Price = Soup.find_all(class_="_30jeq3 _16Jk6d")
Price[0].get_text()
```

(2) Get All The Customers Names In This HTML Page:-

```python
names = Soup.find_all(class_="_2sc7ZR _2V5EHH")
names=list(names)
print(type(names))
print(names)

len(names)
```

(3) Get All The Customers Names In List Form:-

```python
cust_name = []
for i in names:
    cust_name.append(i.get_text())
cust_name
```

(4) Get All The Customers Comments On This Product:-

```python
Comment = Soup.find_all(class_="_2-N8zT")
Comment
```

(5) Get All The Customers Comments In List Form:-

```python
Review_Comment = []
for i in Comment:
    Review_Comment.append(i.get_text())
Review_Comment
```

(6) Check How Much STARS Gives Customers To This Product:-
            
            (STARS RATING OUT OF 5 STARS)

```python
Stars = Soup.find_all("div",class_="_3LWZlK _1BLPMq")
Stars
```

(7) Check How Much STARS Gives Customers To This Product In List Form:-
                
                  (STARS RATING OUT OF 5 STARS)

```python
Review_Stars = []
for i in range(0,len(Stars)):
    Review_Stars.append(Stars[i].get_text())
Review_Stars
```

(8) Get All The Customers Review To This Product:-

```python
Review = Soup.find_all("div",class_="t-ZTKy")
Review
```

(9) Get All The Customers Review To This Product In List Form:-

```python
Content_Review = []
for i in range(0,len(Review)):
    Content_Review.append(Review[i].get_text())
Content_Review
```

(10) Get All The Likes And Dislikes To Customers Review:-

```python
LikesDislikes = Soup.find_all(class_="_3c3Px5")
LikesDislikes
```

(11) Get All The Likes And Dislikes To Customers Review In List Form:-

```python
Product_LikesDislikes = []
for i in range(10,len(LikesDislikes)):
    Product_LikesDislikes.append(LikesDislikes[i].get_text())
Product_LikesDislikes
```

```python
like=[]
dislike=[]
for i in range(10,len(Product_LikesDislikes)):
    print(i)
    if i%2==0:
        like.append(Product_LikesDislikes[i])
    elif i%2!=0:
        dislike.append(Product_LikesDislikes[i]) 
        
print(like)
print(dislike)

    
```
